```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
               # includes math scores and cs scores.find relationship between math and cs scores

# load data    
df = pd.read_csv('test_scores.csv');
```

#### func for gradient descent

```python

def gradient_descent(x,y):
    # starting parameter values
    m_curr = b_curr = 0

    learning_rate = 0.0002
    m = len(x)
    plt.scatter(x, y, color='red', marker='o')

    for i in range(10000):
        y_predict = m_curr * x + b_curr
        plt.plot(x, y_predict, color='green')
        
        # calculate derivatives
        m_derv = -(2/m)*sum(x*(y-y_predict))
        b_derv = -(2/m)*sum(y-y_predict)

        # update parameters
        m_curr = m_curr - learning_rate * m_derv
        b_curr = b_curr - learning_rate * b_derv
        
```

```python
x = np.array(df['math'])
y = np.array(df['cs'])
```

```python
gradient_descent(x, y)
```
